```python
from PIL import Image

def resize_img(path):
    print(path)
    im = Image.open(path)
    im = im.resize((768, 1024), Image.BICUBIC)
    im.save(path)
```

```python
for path in os.listdir('/content/fashion-eye-try-on/VITON-HD/datasets/test/image/'):
    resize_img(f'/content/fashion-eye-try-on/VITON-HD/datasets/test/image/{path}')
for path in os.listdir('/content/fashion-eye-try-on/VITON-HD/datasets/test/cloth/'):
    resize_img(f'/content/fashion-eye-try-on/VITON-HD/datasets/test/cloth/{path}')
for path in os.listdir('/content/fashion-eye-try-on/VITON-HD/datasets/test/cloth-mask/'):
    resize_img(f'/content/fashion-eye-try-on/VITON-HD/datasets/test/cloth-mask/{path}')
```
